```python
import itertools
import numpy as np
from matplotlib import pyplot as plt
```

Part 3 - Question 1

```python
pts_x = [1, 1, 2]
pts_y = [1, 2, 2]

axs_3_1 = plt.subplots(nrows=2, ncols=4, sharex=True, sharey=True, figsize=[10, 4])[1].flat
colors = ['b', 'r']
bin_strings = ["".join(seq) for seq in itertools.product("01", repeat=3)]

for i in range(8):
    b_str =  bin_strings[i]
    c = [colors[int(b_str[j])] for j in range(len(b_str))]
    axs_3_1[i].scatter(pts_x, pts_y, c=c)
    
plt.xlim(-1, 3); plt.ylim(-1, 3)
plt.show()
```

Part 3 - Question 2

```python
pts_x = [1, 1.5, 2, 1.5]
pts_y = [1.5, 2, 1.5, 1]

axs_3_2a1 = plt.subplots(nrows=4, ncols=4, sharex=True, sharey=True, squeeze=True, figsize=[10, 10])[1].flat
colors = ['b', 'r']
bin_strings = ["".join(seq) for seq in itertools.product("01", repeat=4)]

for i in range(16):
    b_str =  bin_strings[i]
    c = [colors[int(b_str[j])] for j in range(len(b_str))]
    axs_3_2a1[i].scatter(pts_x, pts_y, c=c)
    
plt.xlim(0, 3); plt.ylim(0, 3)
plt.show()
```
